```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

page = 55

while True:
    url = "https://tabelog.com/tw/tokyo/rstLst/"+str(page)+"/"
    response = urlopen(url)
    html = BeautifulSoup(response)
    # print(html)
    for r in html.find_all("li" , class_ = "list-rst js-list-item"):
        ja = r.find("small" , class_ = "list-rst__name-ja")
        en = r.find("a" , class_="list-rst__name-main js-detail-anchor")
        ratings = r.find("b" , class_ = "c-rating__val")
        #萃取紙條(.text)   萃取特徵["特徵"]
        print(ratings.text,ja.text , en.text , en["href"])
    page = page + 1
```
